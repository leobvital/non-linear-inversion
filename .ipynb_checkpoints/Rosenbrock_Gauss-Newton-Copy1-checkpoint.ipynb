# Método de Gauss-Newton aplicado à função Rosenbrock

Este algoritmo aplica o método de otimização Steepest descent à função matemática de Rosenbrock, a qual é definida pela equação geral:

\begin{equation}
f(x,y) = (a - x)^2 + b(y - x^2)^2 ,
\end{equation}

onde $a$ e $b$ são constantes e o mínimo dessa função é $(x,y)=(a,a^2)$, onde $f(a,a^2)=0$. Normalmente $a$ e $b$ são iguais a 1 e a 100, respectivamente, nesse caso, o mínimo dessa função é $(x,y)=(1,1)$.

Então a função a ser minimizada através do método steepest descente é:

\begin{equation}
f(p_1,p_2) = 100(p_2 - p_1^2)^2 + (p_1 - 1)^2 ,
\end{equation}

A proposta do código é encontrar o mínimo da função através de muitas iterações.

# Corrigir para o Gauss-Newton

O método de Newton é um método de gradiente aceitável, o qual expande a função objetivo em série de Taylor até segunda ordem em torno de $\mathbf{p}_i$:

\begin{equation}
\phi(\mathbf{p}) = \phi(\mathbf{p}_i) + \mathbf{J}_i^T(\mathbf{p} - \mathbf{p}_i) + \frac{1}{2}(\mathbf{p} - \mathbf{p}_i)^T(\mathbf{A}^T_i \mathbf{A}_i)^{-1}(\mathbf{p} - \mathbf{p}_i)
\end{equation}

onde $\mathbf{J} = \boldsymbol{\nabla}_{\mathbf{p}}\phi |_{\mathbf{p} =\mathbf{p}_i}$ e $\mathbf{A}$ é a matriz de sensibilidade.

Todo método de gradiente aceitável é definido pela equação:

\begin{equation}
\mathbf{p}_{i+1} = \mathbf{p}_i - \rho_i \mathbf{R}_i\mathbf{J}_i
\end{equation}

No método de Gauss-Newton $\rho_i = 1$ e $\mathbf{R}_i = (\mathbf{A}^T_i \mathbf{A}_i)^{-1}$. Seu algoritmo fica:

1) estabelece um chute inicial $\mathbf{p}_i$

2) calcula $\mathbf{J}_i$ e $(\mathbf{A}^T_i \mathbf{A}_i)^{-1}$

3) calcula $\mathbf{p}_{i+1} = \mathbf{p}_i - (\mathbf{A}^T_i \mathbf{A}_i)^{-1} \mathbf{J}_i$
 
4) testa a convergência, se for satisfeita o algoritmo para, caso contrário retorna para 1)

# Importando pacotes

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors, ticker
from visual import histeq
```

## Plotando a função

```python
#definindo os parâmetros
p1 = np.linspace(-2., 2., 20)
p2 = np.linspace(-3., 3., 20)

#criando um grid
p1, p2 = np.meshgrid(p1, p2)

# calculando a phi
phi = 100.*(p2 - p1*p1)*(p2 - p1*p1) + (p1 - 1)*(p1 - 1)

# equalizando as cores
phi_eq = histeq(phi)
phi_eq = np.reshape(phi_eq,(20,20))

#plot
plt.figure(figsize=(5,5))
plt.plot(1., 1., 'oy')
plt.title('Funcao Rosenbrock', fontsize=18)
plt.contourf(p1, p2, phi_eq, 100, cmap='jet')
plt.xlabel('$p_1$', fontsize=18)
plt.ylabel('$p_2$', fontsize=18)
plt.colorbar(pad=0.01, aspect=20, shrink=1)
plt.savefig('plot_rosenbrock.png', pad_inches=0.0)
plt.show()
```

# Algoritmo Steepest Descent

```python
# calculando a phi
def rosenbrock(p0):
    '''
    Calcula a função de Rosenbrock
    
    entrada
    p0: array 1D - contém duas componentes que podem ser arrays 1D
    
    output
    phi: array 1D - valor da função calculada em cada ponto (p1, p2)
    '''
    phi = 100.*(p0[1] - p0[0]*p0[0])*(p0[1] - p0[0]*p0[0]) + (p0[0] - 1)*(p0[0] - 1)
    
    return phi

# calculando gradiente de phi
def grad_rosenbrock(p0):
    '''
    Calcula o vetor de resíduos da função de Rosenbrock
    
    entrada
    p0: array 1D - contém duas componentes que podem ser arrays 1D
    
    output
    res: array 1D - vetor de resíduos calculado em cada ponto (p1, p2)
    '''
    grad = np.zeros((2))
    grad[0] = -400.*p0[0](p0[1] - p0[0]*p0[0]) 
    grad[1] = 200.*(p2 -p0[0]*p0[0])

    return grad

# calculando a hessiana de phi
def jacobi_rosenbrock(p0):
    '''
    Calcula a Jacobiana da função de Rosenbrock
    
    entrada
    p0: array 1D - contém duas componentes que podem ser arrays 1D
    
    output
    A: array 2D - matriz Jacobiana calculada em cada ponto (p1, p2)
    '''
    A = np.zeros((2,2))
    A[0,0] = -400.*p0[0](p0[1] - p0[0]*p0[0])
    A[0,1] = 200.*(p2 -p0[0]*p0[0])
    A[1,0] = -400.*p0[0]
    A[0,1] = A[1,0]
    
    return A
```

```python
# chute inicial
p01 = 1.8
p02 = 2.5

p0 = np.array([p01,p02])

# parâmetros para o critério de Armijo
alfa = 0.0001
beta = 0.4
m = 0
lamb = 1.

phi0 = rosenbrock(p0)

p1_list = [p01]
p2_list = [p02]
phi_list = []
```

```python
for i in range(10):
    # calculando o valor da função e do gradiente em p0
    phi0 = rosenbrock(p0)
    grad = grad_rosenbrock(p0)
    A = jacobi_rosenbrock(p0)
    ATA = np.dot(A.T,A)
    
    # calculando p e atualizando para a próxima iteração
    p = p0 - np.linalg.solve(ATA, np.dot(A.T,res))
    p0 = p
    
    # armazenando os resultados em listas para o plot
    phi_list.append(phi0)
    p1_list.append(p0[0])
    p2_list.append(p0[1])
    
    # printa os valores
    print 'vetor de parametros', p0
    #print 'valor da funcao', phi0
    #print 'vetor gradiente', grad
```

```python
#plot
plt.figure(figsize=(5,5))
plt.plot(1., 1., 'oy')
plt.plot(p1_list, p2_list, '.w-', markersize=8)
plt.title('GN - Rosenbrock', fontsize=18)
plt.contourf(p1, p2, phi_eq, 40, cmap='jet')
plt.xlabel('$p_1$', fontsize=18)
plt.ylabel('$p_2$', fontsize=18)
plt.colorbar(pad=0.01, aspect=20, shrink=1)
plt.savefig('GN_rosenbrock.png', pad_inches=0.0)
plt.show()
```

```python
plt.title('Convergence', fontsize = 18)
plt.xlabel('Iteration number')
plt.ylabel('Phi value')
plt.plot(phi_list, 'o')
plt.savefig('convergence.png', dpi=300)
plt.show()
```
